```python
from leap_hand import LeapHand

# PyMJCF
from dm_control import mjcf

import PIL.Image
```

```python
hand = LeapHand()
print(hand.model.to_xml_string())
```

```python
# Instantiate the physics and render.
physics = mjcf.Physics.from_mjcf_model(hand.model)
PIL.Image.fromarray(physics.render())
```
